```python
import matplotlib.pyplot as plt
```

```python
# Criando uma Classe
class Campanha:
  def __init__(self, canal, investimento, cliques, conversoes):
    self.canal = canal
    self.investimento = investimento
    self.cliques = cliques
    self.conversoes = conversoes

  # Criando um Método de "Custo por Clique"
  # Retorna o investimento por clique
  def cpc(self):
    return self.investimento / self.cliques
```

```python
# Criando uma lista de campanhas com a Classe Campanha
campanhas = [
    Campanha('Facebook Ads', 1000, 15000, 150),
    Campanha('Google Ads', 1200, 10000, 200),
    Campanha('Email Ads', 500, 5000, 50),
    Campanha('Instagran Ads', 800, 12000, 80),
]
print(campanhas)
```

```python
# Criando uma lista "canais", apenas com a coluna/atributo "canal" a partir de uma outra lista "campanhas"
canais = [c.canal for c in campanhas]
canais
```

```python
# Criando uma lista com o resultado do método cpc aplicado em cada uma das campanhas (Custo por Clique de cada campanha)
cpcs = [c.cpc() for c in campanhas]
cpcs
```

```python
# Criando e formatando o gráfico
plt.bar(canais, cpcs, color='Red')
plt.title('Custos por Clique')
plt.xlabel('Canais')
plt.ylabel('Custo em Reais R$')
plt.grid(True)
plt.show()
```
